```python
import os
import duckdb
import pandas as pd
from dotenv import load_dotenv
from mktools.get_data import load_data_pd

load_dotenv()

DUCKDB_PATH = r"C:\Users\Cooper\dbs\duckdb"
```

```python
df = load_data_pd(
    sheet_name="data_main",
    sheet_id=os.environ["SHEET_ID"],
    usecols=[
        "UID",
        "SUID",
        "NAME",
        "CHARACTER",
        "MAP",
        "PLACE",
        "PLAYERS",
        "DATE",
        "SEASON",
    ],
)

df
```

```python
df["DATE"] = pd.to_datetime(
    df["DATE"],
    format="%Y-%m-%d %H:%M:%S"
)

df
```

```python
with duckdb.connect(rf"{DUCKDB_PATH}\mkdata.db") as con:

    con.sql('CREATE OR REPLACE TABLE data_main AS FROM df')
```

```python
c = duckdb.connect(rf"{DUCKDB_PATH}\mkdata.db")

c.table("data_main")
```

```python
c.table("data_main").df().tail()
```
